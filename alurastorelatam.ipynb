### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



###**Ingreso total por cada tienda**

####En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
Ingreso_tienda1 =tienda['Precio'].sum()
print(f"La tienda 1 su ingreso total es: {Ingreso_tienda1}.")
```

```python
Ingreso_tienda2 =tienda2['Precio'].sum()
print(f"La tienda 2 su ingreso total es: {Ingreso_tienda2}.")
```

```python
Ingreso_tienda3 =tienda3['Precio'].sum()
print(f"La tienda 3 su ingreso total es: {Ingreso_tienda3}.")
```

```python
Ingreso_tienda4 =tienda4['Precio'].sum()
print(f"La tienda 4 su ingreso total es: {Ingreso_tienda4}.")
```

##Ingreso total de cada tienda

```python
# Usar un diccionario o lista para evitar repetición
tiendas = [tienda, tienda2, tienda3, tienda4]
ingresos = [t['Precio'].sum() for t in tiendas]

# Convertir a números
ingresos = [float(ingreso) for ingreso in ingresos]

# Imprimir resultados
for i, ingreso in enumerate(ingresos, 1):
    print(f"La tienda {i} su ingreso total es: ${ingreso:,.2f}")
```

# 2. Ventas por categoría

##**Ventas por categoría**

####**En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.**

```python
Ventas_por_categoria1 = tienda.groupby('Categoría del Producto')['Producto'].count()
```

```python
Ventas_por_categoria4 = tienda4.groupby('Categoría del Producto')['Producto'].count()
print(Ventas_por_categoria4)
```

# 3. Calificación promedio de la tienda

```python
import matplotlib.pyplot as plt

# Calcular calificaciones promedio para todas las tiendas
calificaciones = []
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Calcular promedio de cada tienda
promedio_tienda1 = round(tienda['Calificación'].mean(), 2)
promedio_tienda2 = round(tienda2['Calificación'].mean(), 2)
promedio_tienda3 = round(tienda3['Calificación'].mean(), 2)
promedio_tienda4 = round(tienda4['Calificación'].mean(), 2)

calificaciones = [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]

# Imprimir valores
for i, (nombre, calif) in enumerate(zip(nombres_tiendas, calificaciones), 1):
    print(f"{nombre}: {calif}")
```

####Ingresos totales por Tienda

```python
# Importar librerias
import pandas as pd
import matplotlib.pyplot as plt

# Crear DataFrame con los resultados
df_ingresos = pd.DataFrame({
    'Tienda': [f'Tienda {i}' for i in range(1, len(ingresos) + 1)],
    'Ingreso': ingresos
})

# Graficar usando pandas
ax = df_ingresos.plot(kind='bar', x='Tienda', y='Ingreso',
                      figsize=(10, 6), color='teal', legend=False)
ax.set_title('Ingresos Totales por Tienda', fontsize=14)
ax.set_ylabel('Ingreso ($)', fontsize=12)

# Añadir valores
for p in ax.patches:
    ax.annotate(f'${p.get_height():,.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

###Venta por categoria: Por tienda

```python
categorias = [
    'Artículos para el hogar',
    'Deportes y diversión',
    'Electrodomésticos',
    'Electrónicos',
    'Instrumentos musicales',
    'Juguetes',
    'Libros',
    'Muebles'
]

ventas_tienda1 = Ventas_por_categoria1.reindex(categorias, fill_value=0)
ventas_tienda2 = Ventas_por_categoria2.reindex(categorias, fill_value=0)
ventas_tienda3 = Ventas_por_categoria3.reindex(categorias, fill_value=0)
ventas_tienda4 = Ventas_por_categoria4.reindex(categorias, fill_value=0)

## import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(categorias))
ancho = 0.2

plt.figure(figsize=(14, 6))

plt.bar(x - 1.5*ancho, ventas_tienda1, width=ancho, label='Tienda 1')
plt.bar(x - 0.5*ancho, ventas_tienda2, width=ancho, label='Tienda 2')
plt.bar(x + 0.5*ancho, ventas_tienda3, width=ancho, label='Tienda 3')
plt.bar(x + 1.5*ancho, ventas_tienda4, width=ancho, label='Tienda 4')

plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Productos Vendidos')
plt.title('Comparación de Ventas por Categoría y Tienda')
plt.xticks(x, categorias, rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Obtener todas las categorías únicas
categorias_todas = pd.concat([tienda, tienda2, tienda3, tienda4])['Categoría del Producto'].unique()
categorias_todas.sort()

# Crear DataFrame con todas las categorías y tiendas
datos_comparativos = pd.DataFrame(index=categorias_todas)

# Asignar los datos de cada tienda al DataFrame
datos_comparativos['Tienda 1'] = Ventas_por_categoria1
datos_comparativos['Tienda 2'] = Ventas_por_categoria2
datos_comparativos['Tienda 3'] = Ventas_por_categoria3
datos_comparativos['Tienda 4'] = Ventas_por_categoria4

# Rellenar valores NaN con 0
datos_comparativos = datos_comparativos.fillna(0)

print("\n=== TABLA COMPARATIVA COMPLETA ===")
print(datos_comparativos)
print("\n=== TOTAL POR TIENDA ===")
print(datos_comparativos.sum())

# ============================================
# 3. GRÁFICA DE BARRAS AGRUPADAS
# ============================================

# Crear figura
fig, ax = plt.subplots(figsize=(12, 7))

# Posiciones en el eje X
x = np.arange(len(categorias_todas))
ancho = 0.18  # Ancho de cada barra

# Colores para cada tienda
colores = ['#2E86C1', '#E74C3C', '#28B463', '#F39C12']  # Azul, Rojo, Verde, Naranja

# Crear barras para cada tienda
for i in range(4):
    tienda_nombre = f'Tienda {i+1}'
    posicion = x + (i - 1.5) * ancho

    ax.bar(posicion,
           datos_comparativos[tienda_nombre],
           width=ancho,
           label=tienda_nombre,
           color=colores[i],
           alpha=0.85,
           edgecolor='black',
           linewidth=1)

# Personalizar la gráfica
ax.set_xlabel('Categorías de Productos', fontsize=12, fontweight='bold')
ax.set_ylabel('Cantidad de Productos', fontsize=12, fontweight='bold')
ax.set_title('Comparación de Productos por Categoría en 4 Tiendas',
             fontsize=14, fontweight='bold', pad=20)

# Configurar eje X
ax.set_xticks(x)
ax.set_xticklabels(categorias_todas, rotation=45, ha='right', fontsize=10)

# Añadir leyenda
ax.legend(title='Tiendas', fontsize=10)

# Añadir grid
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)  # Grid detrás de las barras

# Ajustar límites
max_valor = datos_comparativos.values.max()
ax.set_ylim(0, max_valor * 1.15)

# Mejorar estética
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
